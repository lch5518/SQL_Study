### 등가 조인(JOIN)



* JOIN 이라는 단어가 들어가지 않은 비 표준
* JOIN에 관한 표준이 정해지기 전부터 사용되어왔다.

```python
%sql select * from employees;
```

```python
%sql select * from departments;
```

```python
%%sql 
SELECT e.employee_id, d.department_name 
FROM   employees e, departments d
WHERE  e.department_id = d.department_id;
```

### NATURAL JOIN



* NATURAL JOIN은 두 테이블에서 동일한 이름을 가진 모든 열을 기준으로 한다.
* 두 테이블의 일치하는 모든 열에서 같은 값을 가진 행을 선택한다.
* 동일한 이름을 가진 열의 데이터 유형이 다를 경우 오류가 반환된다.
<br><br>

```python
%%sql 
SELECT employee_id, department_name, department_id 
FROM   employees 
NATURAL JOIN departments;
```

```python
%%sql 
SELECT department_id, department_name, location_id, city 
FROM   departments 
NATURAL JOIN locations;
```

###  JOIN ... USING (...) 



* 여러 열이 같은 이름을 같은 이름을 가질 때, 한 열만을 일치 시킬 수 있다.
* USING 절에 테이블 이름이나, 테이블 이름 별칭을 사용해서는 안된다.
* USING(...)에서 괄호를 생략하면 에러
* NATURAL JOIN 에 USING을 사용하면 안된다.

```python
%%sql 
SELECT employee_id, department_name, department_id
FROM   employees JOIN departments 
USING  (department_id);
```

```python
%%sql 
SELECT employee_id, department_name, department_id
FROM   employees JOIN departments 
USING  department_id;
```

###  JOIN ... ON 



* NATURAL JOIN 이나 JOIN ... USING (...) 은 두 테이블에서 같은 이름을 가진 열이 존재할 때 같은 값을 가지는 경우만을 검색할 수 있다.
* JOIN ... ON 은 임의의 조건을 지정할 때 사용한다.

```python
%%sql 
SELECT e.employee_id, d.department_name, d.department_id
FROM   employees e JOIN departments d
ON     e.department_id = d.department_id;
```

### 추가적인 조건

```python
%%sql
SELECT e.employee_id, e.last_name, e.department_id, 
       d.department_id, d.location_id
FROM   employees e JOIN departments d
ON     e.department_id = d.department_id
AND    e.manager_id = 149 ;
```

### 3개 이상의 테이블  JOIN

```python
%%sql 
SELECT employee_id, city, department_name
FROM   employees e 
JOIN   departments d
ON     e.department_id = d.department_id
JOIN   locations l
ON     d.location_id = l.location_id;
```

### 셀프 JOIN

```python
%%sql
SELECT e.last_name emp, m.last_name mgr
FROM   employees e JOIN employees m
ON     e.manager_id = m.employee_id;
```

### INNER JOIN&nbsp;&nbsp;&nbsp; vs. &nbsp;&nbsp;&nbsp; OUTER JOIN



* INNER JOIN 이란 두 테이블을 조인하여 일치하는 행만 반환한다. 지금까지 우리는 INNER JOIN 에 해당하는 사용법만 살펴 보았다.
* OUTER JOIN 이란 두 테이블을 조인하여 일치하는 행과 함께 일치하지 않는 왼쪽(또는 오른쪽) 테이블의 행도 반환한다.


### LEFT OUTER JOIN


```python
%%sql
SELECT last_name, d.department_id, department_name
FROM   employees e LEFT OUTER JOIN departments d
ON     e.department_id = d.department_id ;
```

### RIGHT OUTER JOIN


```python
%%sql
SELECT e.last_name, e.department_id, d.department_name
FROM   employees e RIGHT OUTER JOIN departments d
ON     e.department_id = d.department_id ;
```

### CROSS JOIN


* 두 테이블의 카치안 곱을 생성한다.

```python
%%sql
SELECT last_name, department_name
FROM   employees
CROSS JOIN departments ;
```

#### 카티시안(Cartesian) 곱<span style="color:red";> (대부분 오류) </span>



![](pic/pic5_2.jpg)
<br><br>
카티시안 곱은 다음 경우 생성된다.

* JOIN 조건을 생략한 경우
* JOIN 조건이 부적합한 경우
* 첫번째 table의 모든 행이 두번째 table의 모든 행에 JOIN된 경우

카티시안 곱이 생성되지 않도록 WHERE 절에 유효한 JOIN 조건을 포함시켜야 한다.

```python
%%sql
SELECT last_name, department_name
FROM   employees JOIN departments ;
```

<span style="color:red">[문제 1]</span>   
모든 사원의 이름, 부서번호, 부서이름을 표시하라.

```python
%%sql 
SELECT last_name, department_id, department_name
FROM   employees 
NATURAL JOIN departments;
```

<br><br>
<span style="color:red">[문제 2]</span>   
부서 80에 속하는 모든 업무의 고유목록과 부서위치 ID를 나타내라.

<span style="color:orange">
결과는 2줄  

```python
%%sql
select distinct job_id, location_id, d.department_id
from employees e join departments d
on d.department_id = 80 and e.department_id = d.department_id
```

<br><br>
<span style="color:red">[문제 3]</span>   
커미션을 받는 모든 사원의 이름, 부서이름, 위치 ID 및 도시를 표시하라.

```python
%%sql
select e.last_name, d.department_name, location_id
from employees e join departments d
on e.commission_pct is not null and e.department_id = d.department_id
```

<br><br>
<span style="color:red">[문제 4]</span>   
이름(last_name)에 a 가 포함된 모든 사원의 이름, 부서이름을 나타내라.

```python
%%sql
select e.last_name, d.department_name
from employees e join departments d
on e.last_name like '%a%' and e.department_id = d.department_id
```

<br><br>
<span style="color:red">[문제 5]</span>   
Toronto에서 근무하는 모든 사원의 이름, 업무, 부서번호, 부서이름을 표시하라.   

<span style="color:orange">
결과는 2줄 

```python
%%sql 
SELECT city, e.last_name, e.job_id, d.department_id, department_name
FROM   employees e 
JOIN   departments d
ON     e.department_id = d.department_id
JOIN   locations l
ON     d.location_id = l.location_id
and city = 'Toronto'
```

<br><br>
<span style="color:red">[문제 6]</span>   
사원의 이름 및 사원번호를 관리자 이름 및 관리자 번호와 함께 나타내라.

```python
%%sql
SELECT e.last_name 사원이름, e.employee_id 사원번호, m.last_name 관리자이름, m.employee_id 관리자번호
FROM   employees e JOIN employees m
ON     e.manager_id = m.employee_id;
```

<br><br>
<span style="color:red">[문제 7]</span>   
문제 6을 수정하여, 관리자가 없는 경우까지 나타나도록 하라.

```python
%%sql
SELECT e.last_name 사원이름, e.employee_id 사원번호, m.last_name 관리자이름, m.employee_id 관리자번호
FROM   employees e left outer JOIN employees m
ON     e.manager_id = m.employee_id;
```
